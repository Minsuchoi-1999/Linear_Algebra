## 2.2 The Idea of Elimination

이 Chapter에서는 Elimination에 대한 개념과 방법에 대해 나옵니다.
우선은 Permutation 없이 Pivot이 0이 나오면 Elimination을 멈추고 Failure이라고 하도록 짜겠습니다. 

```python
import sys

def Elimination1(matrix):
    for i in range(len(matrix)):
        pivot = matrix[i][i] ## pivot을 구한다.
        
        if pivot == 0:
            print(f"{i+1}행 {i+1}열의 pivot이 0입니다!")
            return False
            #pivot이 0이면 False를 반환해 Failure을 출력한다.
        else:
            for j in range(i+1, len(matrix)):
                multiplier = matrix[j][i]/pivot
                
                #각 행의 pivot에 대한 multiplier를 구한다
                
                for k in range(0, len(matrix[0])):
                    matrix[j][k] -= (matrix[i][k] * multiplier)
                    #multiplier를 이용해서 matrix의 각 행을 Elimination해준다.
        
    return True
            #성공적으로 Elimination을 했다면 True를 반환해 Elimination된 Matrix를 출력한다
    
def print_matrix(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    
    for i in range(rows):
        for j in range(cols):
            print(matrix[i][j], end ="\t")
        print()
    
matrix = list()

rows = int(input("Matrix A 행 수를 입력하세요: "))
cols = int(input("Matrix A 열 수를 입력하세요: "))

for i in range(rows):
        row = input(f"행 {i+1}을 공백으로 구분하여 입력하세요: ").split()
        if len(row) != cols:
            print("열 수와 입력된 값의 개수가 일치하지 않습니다.")
            sys.exit(1)
        else:
            matrix.append([float(val) for val in row])
print()

if(Elimination1(matrix)):
    print_matrix(matrix)
else:
    print("Failure")
```

위의 프로그램에서 조금만 변형하면 Back substitution을 통해 해를 구할 수도 있습니다.
일단은 Square Augmented matrix를 입력받도록 변형해서 unique solution까지 구하도록 변형해보겠습니다.

## 2.3 Elimination Using Matrices

여기서는 Elimination Matrix를 직접 구합니다.
이를 통해 EA=U 꼴의 출력이 가능합니다. 이 역시 Elimination1에서 조금의 변화만 주면 직접 구할 수 있습니다.
Permutaion Matrix에 대한 이야기도 나오지만, 다음 장에 본격적으로 다룰 예정입니다.

```python
import sys

def print_matrix(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    
    for i in range(rows):
        for j in range(cols):
            print(matrix[i][j], end ="\t")
        print()
        
def make_identity(rows, cols):#IA = A 에서 EA = U의 E를 만들기 위해 Identity matrix를 우선 만든다.
    Identity = list()
    for i in range(rows):
        row_Identity = list()
        
        for j in range(cols):
            if(i != j):
                row_Identity.append(0)
            else:
                row_Identity.append(1) # 대각행렬에는 1, 아니면 0을 입력하여 반환한다.
        
        Identity.append(row_Identity)

    return Identity

Elimination_matrices = list()
#각 단계에서의 Elimination matrice를 저장한다. 

def Elimination2(matrix):
    for i in range(len(matrix)):
        pivot = matrix[i][i] ## pivot을 구한다.
        
        if pivot == 0:
            print(f"{i+1}행 {i+1}열의 pivot이 0입니다!")
            return False
            #pivot이 0이면 False를 반환해 Failure을 출력한다.
        else:
            for j in range(i+1, len(matrix)):
                multiplier = matrix[j][i]/pivot
                
                Elimination_Matrix = make_identity(len(matrix), len(matrix[0])) ##Identity Matrix를 만들어 Elimination Matrix를 만든다
                Elimination_Matrix[j][i] = -multiplier
                
                #각 행의 pivot에 대한 multiplier를 구한다
                #Elimination Matrice를 만들기 위해 -multiplier를 저장한다.
                
                Elimination_matrices.append((j+1, i+1, Elimination_Matrix))
                #각 단계의 Elimination Matrice들을 저장한다
                
                for k in range(0, len(matrix[0])):
                    matrix[j][k] -= (matrix[i][k] * multiplier)
                    #multiplier를 이용해서 matrix의 각 행을 Elimination해준다.
        
    return True
            #성공적으로 Elimination을 했다면 True를 반환해 Elimination된 Matrix를 출력한다
    
matrix = list()

rows = int(input("Matrix A 행 수를 입력하세요: "))
cols = int(input("Matrix A 열 수를 입력하세요: "))

original_matrix = list()

for i in range(rows):
        row = input(f"행 {i+1}을 공백으로 구분하여 입력하세요: ").split()
        if len(row) != cols:
            print("열 수와 입력된 값의 개수가 일치하지 않습니다.")
            sys.exit(1)
        else:
            matrix.append([float(val) for val in row])
            original_matrix.append([float(val) for val in row])
print()

if(Elimination2(matrix)):
    for i in range(len(Elimination_matrices)):
        print(f"E{Elimination_matrices[i][0]}{Elimination_matrices[i][1]}:")
        print_matrix(Elimination_matrices[i][2])
        print()
    
    print("A")
    print_matrix(original_matrix)
    print()
    
    print("U")
    print_matrix(matrix)
    print()
else:
    print("No solution")
```

## 2.4 Rules for Matrix Operations

이번 chapter는 2개의 Matrix의 곱을 하는 방법에 대해 다룹니다.
4가지 방법이 있지만 내적을 이용한 방법을 다루겠습니다.

```python
def Matrice_multiplication(Ma, Mb):
    if(len(Ma[0])!=len(Mb)):
        print("Invalid Multiplication")
    else:
        answer = list()
        
        for i in range(len(Ma)):
            temp = [0.0]*len(Mb[0])
            answer.append(temp)
            
        #answer을 위한 Matrix를 초기화한다.
        
        for i in range(len(Ma)):
            for j in range(len(Mb[0])):
                temp = 0
                for k in range(len(Mb)):
                    temp += Ma[i][k]*Mb[k][j]#A의 행, B열의 열을 곱하는 가장 기초적인 방법
                answer[i][j] = temp
                
        return answer


def print_matrix(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    
    for i in range(rows):
        for j in range(cols):
            print(matrix[i][j], end ="\t")
        print()

matrixA = list()

rows = int(input("Matrix A 행 수를 입력하세요: "))
cols = int(input("Matrix A 열 수를 입력하세요: "))

for i in range(rows):
        row = input(f"행 {i+1}을 공백으로 구분하여 입력하세요: ").split()
        if len(row) != cols:
            print("열 수와 입력된 값의 개수가 일치하지 않습니다.")
            sys.exit(1)
        else:
            matrixA.append([float(val) for val in row])
print()

matrixB = list()

rows = int(input("Matrix B 행 수를 입력하세요: "))
cols = int(input("Matrix B 열 수를 입력하세요: "))

for i in range(rows):
        row = input(f"행 {i+1}을 공백으로 구분하여 입력하세요: ").split()
        if len(row) != cols:
            print("열 수와 입력된 값의 개수가 일치하지 않습니다.")
            sys.exit(1)
        else:
            matrixB.append([float(val) for val in row])
print()

print("matrix A:")
print_matrix(matrixA)
print()

print("matrix B:")
print_matrix(matrixB)
print()

print("AB")
print_matrix(Matrice_multiplication(matrixA, matrixB))
```

## 2.5 Inverse Matrices

이번 chapter에서는 Inverse Matrix와 그 값을 구하는 방법에 대해 배웁니다.
가장 기초적인 2 by 2 Matrix 공식부터 Gauss-Jordan Elimination까지 구현해보겠습니다.

```python
def inverse(Ma):
    a = Ma[0][0]
    b = Ma[0][1]
    c = Ma[1][0]
    d = Ma[1][1]
    
    if (a*d-b*c) == 0:
        return None
    else:
        det = a*d-b*c
        answer = [[d/det, -b/det],[-c/det, a/det]]
        return answer
    
def print_matrix(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    
    for i in range(rows):
        for j in range(cols):
            print(matrix[i][j], end ="\t")
        print()

matrixA = list()

rows = int(input("Matrix A 행 수를 입력하세요: "))
cols = int(input("Matrix A 열 수를 입력하세요: "))

for i in range(rows):
        row = input(f"행 {i+1}을 공백으로 구분하여 입력하세요: ").split()
        if len(row) != cols:
            print("열 수와 입력된 값의 개수가 일치하지 않습니다.")
            sys.exit(1)
        else:
            matrixA.append([float(val) for val in row])
print()

print("matrix A:")
print_matrix(matrixA)
print()

print("matrix inverse of A:")

matrix_inverse = inverse(matrixA)

if matrix_inverse:
    print_matrix(matrix_inverse)
else:
    print("There is no inverse")
print()
```

이번에는 Gauss-Jordan Elimination에 대해서도 구현해보겠습니다.
전체적인 구조는 Elimination1에서 구현한 함수와 비슷하나, 맨 마지막 행에서 첫 행으로 거꾸로 빼주어야 하고 대각성분은 1로 만들어주어야합니다.

```python
import sys

def Gauss_Jordan(matrix, matrix2):
    for i in range(len(matrix)):
        pivot = matrix[i][i] ## pivot을 구한다.
        
        if pivot == 0:
            print(f"{i+1}행 {i+1}열의 pivot이 0입니다!")
            return None
            #pivot이 0이면 False를 반환해 Failure을 출력한다.
        else:
            for j in range(i+1, len(matrix)):
                multiplier = matrix[j][i]/pivot
                
                #각 행의 pivot에 대한 multiplier를 구한다
                
                for k in range(0, len(matrix[0])):
                    matrix[j][k] -= (matrix[i][k] * multiplier)
                    matrix2[j][k] -= (matrix2[i][k] * multiplier)
                    #multiplier를 이용해서 matrix, matrix2의 각 행을 Elimination해준다.
                    
    #여기서 끝나지 않고 아래에서 위로 한번더 빼준다.
    for i in range(len(matrix)-1, -1 , -1):
        pivot = matrix[i][i] ## pivot을 구한다.
        
        for j in range(i-1, -1, -1):
            multiplier = matrix[j][i]/pivot
                
            #각 행의 pivot에 대한 multiplier를 구한다
                
            for k in range(0, len(matrix[0])):
                matrix[j][k] -= (matrix[i][k] * multiplier)
                matrix2[j][k] -= (matrix2[i][k] * multiplier)
                #multiplier를 이용해서 matrix, matrix2의 각 행을 Elimination해준다.   
        
    #대각성분을 1로 만들어준다.
    for i in range(len(matrix)):
        pivot = matrix[i][i]
        matrix[i][i] /= pivot
            
        for j in range(len(matrix)):
            matrix2[i][j]/= pivot
        
    return matrix2
            #성공적으로 Elimination을 했다면 True를 반환해 Elimination된 Matrix를 출력한다

def make_identity(rows, cols): #Gauss-Jordan Elimination을 위해 Identity matrix를 만든다.
    Identity = list()
    for i in range(rows):
        row_Identity = list()
        
        for j in range(cols):
            if(i != j):
                row_Identity.append(0)
            else:
                row_Identity.append(1) # 대각행렬에는 1, 아니면 0을 입력하여 반환한다.
        
        Identity.append(row_Identity)

    return Identity

    
def print_matrix(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    
    for i in range(rows):
        for j in range(cols):
            print(matrix[i][j], end ="\t")
        print()
    
matrix = list()

rows = int(input("Matrix A 행 수를 입력하세요: "))
cols = int(input("Matrix A 열 수를 입력하세요: "))

for i in range(rows):
        row = input(f"행 {i+1}을 공백으로 구분하여 입력하세요: ").split()
        if len(row) != cols:
            print("열 수와 입력된 값의 개수가 일치하지 않습니다.")
            sys.exit(1)
        else:
            matrix.append([float(val) for val in row])
print()

Identity_matrix = make_identity(rows, cols)
answer = Gauss_Jordan(matrix, Identity_matrix)

if answer:
    print_matrix(answer)
else:
    print("Failure")
```

## 2.7 Transposes and Permutations

이번 chapter에서는 Transposes와 Permutation Matrix를 통해 PA=LU를 구합니다.
우선은 Transpose에 대해 구현, 후 Permutation Matrix를 포함한 PA=LU를 구현해보겠습니다.

```python
def transpose(Ma):
    answer = list()
    for _ in range(len(Ma[0])):
        temp = [0.0]*len(Ma)
        answer.append(temp)
        
    for i in range(len(Ma)):
        for j in range(len(Ma[0])):
            answer[j][i] = Ma[i][j]
    return answer

def print_matrix(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    
    for i in range(rows):
        for j in range(cols):
            print(matrix[i][j], end ="\t")
        print()
    
matrix = list()

rows = int(input("Matrix A 행 수를 입력하세요: "))
cols = int(input("Matrix A 열 수를 입력하세요: "))

for i in range(rows):
        row = input(f"행 {i+1}을 공백으로 구분하여 입력하세요: ").split()
        if len(row) != cols:
            print("열 수와 입력된 값의 개수가 일치하지 않습니다.")
            sys.exit(1)
        else:
            matrix.append([float(val) for val in row])
print("Matrix A : ")
print_matrix(matrix)
print()

T_matrix = transpose(matrix)

print("A^T :")
print_matrix(T_matrix)
print()
```

이제는 위에서 다루었던 A=LU에서 중간 과정에서 pivot이 0인 상황을 다루는 PA=LU에 대해 다룹니다.
PA=LU 분해는 교재에 깊게 다루지않아 따로 조사해보니, 각 열의 가장 큰 원소가 있는 행을 위쪽으로 오도록 정렬하는 방법이 있었습니다.
이는 알고리즘으로 구현하면 A를 PA로 전처리 후 위에서 한 것처럼 LU분해를 하면 됩니다.

실제로 P는 n!의 경우의 수를 가지며, 여기서 나오는 P말고도 다른 permutation행렬도 LU분해가 가능하게 바꿀 수 있습니다.
우선은 Greedy하게 정렬하는 하여 PA를 구하는 방법에 대해서만 구현하겠습니다.

```python
import sys

def print_matrix(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    
    for i in range(rows):
        for j in range(cols):
            print(matrix[i][j], end ="\t")
        print()
        
def make_identity(rows, cols):#IA = A 에서 EA = U의 E를 만들기 위해 Identity matrix를 우선 만든다.
    Identity = list()
    for i in range(rows):
        row_Identity = list()
        
        for j in range(cols):
            if(i != j):
                row_Identity.append(0)
            else:
                row_Identity.append(1) # 대각행렬에는 1, 아니면 0을 입력하여 반환한다.
        
        Identity.append(row_Identity)

    return Identity

Elimination_matrices = list()
#각 단계에서의 Elimination matrice를 저장한다. 

def Elimination2(matrix):
    for i in range(len(matrix)):
        pivot = matrix[i][i] ## pivot을 구한다.
        
        if pivot == 0:
            print(f"{i+1}행 {i+1}열의 pivot이 0입니다!")
            return False
            #pivot이 0이면 False를 반환해 Failure을 출력한다.
        else:
            for j in range(i+1, len(matrix)):
                multiplier = matrix[j][i]/pivot
                
                Elimination_Matrix = make_identity(len(matrix), len(matrix[0])) ##Identity Matrix를 만들어 Elimination Matrix를 만든다
                Elimination_Matrix[j][i] = -multiplier
                
                #각 행의 pivot에 대한 multiplier를 구한다
                #Elimination Matrice를 만들기 위해 -multiplier를 저장한다.
                
                Elimination_matrices.append((i+1, j+1, Elimination_Matrix))
                #각 단계의 Elimination Matrice들을 저장한다
                
                for k in range(0, len(matrix[0])):
                    matrix[j][k] -= (matrix[i][k] * multiplier)
                    #multiplier를 이용해서 matrix의 각 행을 Elimination해준다.
        
    return True
            #성공적으로 Elimination을 했다면 True를 반환해 Elimination된 Matrix를 출력한다

def Gauss_Jordan(matrix, matrix2):
    for i in range(len(matrix)):
        pivot = matrix[i][i] ## pivot을 구한다.
        
        if pivot == 0:
            print(f"{i+1}행 {i+1}열의 pivot이 0입니다!")
            return None
            #pivot이 0이면 False를 반환해 Failure을 출력한다.
        else:
            for j in range(i+1, len(matrix)):
                multiplier = matrix[j][i]/pivot
                
                #각 행의 pivot에 대한 multiplier를 구한다
                
                for k in range(0, len(matrix[0])):
                    matrix[j][k] -= (matrix[i][k] * multiplier)
                    matrix2[j][k] -= (matrix2[i][k] * multiplier)
                    #multiplier를 이용해서 matrix, matrix2의 각 행을 Elimination해준다.
                    
    #여기서 끝나지 않고 아래에서 위로 한번더 빼준다.
    for i in range(len(matrix)-1, -1 , -1):
        pivot = matrix[i][i] ## pivot을 구한다.
        
        for j in range(i-1, -1, -1):
            multiplier = matrix[j][i]/pivot
                
            #각 행의 pivot에 대한 multiplier를 구한다
                
            for k in range(0, len(matrix[0])):
                matrix[j][k] -= (matrix[i][k] * multiplier)
                matrix2[j][k] -= (matrix2[i][k] * multiplier)
                #multiplier를 이용해서 matrix, matrix2의 각 행을 Elimination해준다.   
        
    #대각성분을 1로 만들어준다.
    for i in range(len(matrix)):
        pivot = matrix[i][i]
        matrix[i][i] /= pivot
            
        for j in range(len(matrix)):
            matrix2[i][j]/= pivot
        
    return matrix2
            #성공적으로 Elimination을 했다면 True를 반환해 Elimination된 Matrix를 출력한다

def Matrice_multiplication(Ma, Mb):
    if(len(Ma[0])!=len(Mb)):
        print("Invalid Multiplication")
    else:
        answer = list()
        
        for i in range(len(Ma)):
            temp = [0.0]*len(Mb[0])
            answer.append(temp)
            
        #answer을 위한 Matrix를 초기화한다.
        
        for i in range(len(Ma)):
            for j in range(len(Mb[0])):
                temp = 0
                for k in range(len(Mb)):
                    temp += Ma[i][k]*Mb[k][j]#A의 행, B열의 열을 곱하는 가장 기초적인 방법
                answer[i][j] = temp
                
        return answer

def PA(matrix):
    P = make_identity(len(matrix), len(matrix[0]))
    
    for i in range(len(matrix)):
        max_row = i
        for j in range(i + 1, len(matrix)):
            if abs(matrix[j][i]) > abs(matrix[max_row][i]):#절대값이 큰 원소를 위로 정렬하는 방법입니다.
                max_row = j
        P[i], P[max_row] = P[max_row], P[i]
        matrix[i], matrix[max_row] = matrix[max_row], matrix[i]
    return P
    
matrix = list()

rows = int(input("Matrix A 행 수를 입력하세요: "))
cols = int(input("Matrix A 열 수를 입력하세요: "))

original_matrix = list()

for i in range(rows):
        row = input(f"행 {i+1}을 공백으로 구분하여 입력하세요: ").split()
        if len(row) != cols:
            print("열 수와 입력된 값의 개수가 일치하지 않습니다.")
            sys.exit(1)
        else:
            matrix.append([float(val) for val in row])
            original_matrix.append([float(val) for val in row])
print()

P_matrix = PA(matrix)

if(Elimination2(matrix)):
    Elimination_matrices.reverse()
    
    E_matrix = Elimination_matrices[0][2]
    
    for i in range(1, len(Elimination_matrices)):    
        E_matrix = Matrice_multiplication(E_matrix, Elimination_matrices[i][2])
    
    print("E")
    print_matrix(E_matrix)
    print()
    
    Identity = make_identity(rows, cols)
    
    L_matrix = Gauss_Jordan(E_matrix, Identity)
    
    print("L")
    if L_matrix:
        print_matrix(L_matrix)
    else:
        print("There is no L")
    print()
    
    print("P")
    print_matrix(P_matrix)
    print()
    
    print("A")
    print_matrix(original_matrix)
    print()
    
    print("U")
    print_matrix(matrix)
    print()
else:
    print("No solution")
```
